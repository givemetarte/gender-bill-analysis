```python
import pandas as pd 
import numpy as np 
from ast import literal_eval
```

```python
gender = pd.read_csv('13th-gender-bill.csv', encoding='utf-8')
gender = gender.drop('Unnamed: 0', 1)
gender.head(3)
```

## 발의자/제안정당

```python
%time gender['제안자'] = gender['제안자'].apply(literal_eval)
```

```python
def get_proposer(name_list): 
    if len(name_list) > 0: 
        proposer = name_list[0]
    else: 
        proposer = ''
    return proposer
```

```python
%time gender['발의자'] = gender['제안자'].apply(get_proposer)
```

```python
gender['발의자'].unique()
```

```python
minju = ['신영순','김장숙']
sinminju = ['윤재기']
pyunghwa = ['조세형', '박영숙']
tongil = ['신상우','이인제']
```

```python
gender.loc[gender['제안자구분'] == '정부', '제안정당'] = '정부'
gender.loc[gender['제안자구분'] == '정부', '발의자'] = '정부'
gender.loc[gender['제안자구분'] == '위원장', '제안정당'] = '위원장'
gender.loc[gender['제안자구분'] == '위원장', '발의자'] = '위원장'

gender.loc[gender['발의자'].isin(minju), '제안정당'] = '민주정의당'
gender.loc[gender['발의자'].isin(sinminju), '제안정당'] = '신민주공화당'
gender.loc[gender['발의자'].isin(pyunghwa), '제안정당'] = '평화민주당'
gender.loc[gender['발의자'].isin(tongil), '제안정당'] = '통일민주당'
```

## 발의자 성별 구분

```python
women_list = ['이윤자','김장숙','양경자','신영순','도영심','박영숙']

gender.loc[gender['발의자'].isin(women_list), '성별'] = '여성'
gender.loc[(gender['제안자구분'] == '의원') & (~gender['발의자'].isin(women_list)), '성별'] = '남성'
```

```python
gender['성별'].unique()
```

```python
def count_women(name_list, women_list): 
    count = 0 
    for name in name_list: 
        if name in women_list: 
            count += 1
    return count
```

```python
%time gender['여성의원 수'] = gender['제안자'].apply(lambda x: count_women(x, women_list))
```
