```python
import pandas as pd 
import numpy as np 
from ast import literal_eval
```

## 발의자/제안정당

```python
%time gender['제안자'] = gender['제안자'].apply(literal_eval)
```

```python
def get_proposer(name_list): 
    if len(name_list) > 0: 
        proposer = name_list[0]
    else: 
        proposer = ''
    return proposer
```

```python
%time gender['발의자'] = gender['제안자'].apply(get_proposer)
```

## 발의자 성별 구분

```python
women_list = ['박근혜','김영선','전재희','이연숙','김정숙','임진출','손희정',
             '최영희','한명숙','이미경','허운나','김방림','김화중','조배숙',
             '추미애','장영신','김희선','김경천','강숙자']

gender.loc[gender['발의자'].isin(women_list), '성별'] = '여성'
gender.loc[(gender['제안자구분'] == '의원') & (~gender['발의자'].isin(women_list)), '성별'] = '남성'
```

```python
gender['성별'].unique()
```

```python
def count_women(name_list, women_list): 
    count = 0 
    for name in name_list: 
        if name in women_list: 
            count += 1
    return count
```

```python
%time gender['여성의원 수'] = gender['제안자'].apply(lambda x: count_women(x, women_list))
```
