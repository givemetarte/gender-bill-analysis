# 19대 전처리 

- 발의자, 제안정당, 성별 

```python
import pandas as pd 
import numpy as np 
from ast import literal_eval
```

```python
gender = pd.read_csv('19th-gender-bill.csv', encoding='utf-8')
gender = gender.drop('Unnamed: 0', 1)
gender.head(10)
```

## 발의자/제안정당

```python
%time gender['제안자'] = gender['제안자'].apply(literal_eval)
```

```python
def get_proposer(name_list): 
    if len(name_list) > 0: 
        proposer = name_list[0]
    else: 
        proposer = ''
    return proposer
```

```python
%time gender['발의자'] = gender['제안자'].apply(get_proposer)
```

```python
gender[gender['발의자'] == '']['제안자구분'].unique()
```

```python
print(gender['발의자'].unique())
print(len(gender['발의자'].unique()))
```

```python
def get_hangul_name(text):
    text = text.replace('(',' ').replace(')',' ').replace('/',' ').split(' ')
    return text[0].strip()
    
def get_party(text): 
    try: 
        text = text.replace('(',' ').replace(')',' ').replace('/',' ').split(' ')
        return text[1].strip()
    except:
        pass
```

```python
%time gender['발의자_한글'] = gender['발의자'].apply(get_hangul_name)
%time gender['제안정당'] = gender['발의자'].apply(get_party)
```

```python
gender[(gender['제안자구분'] == '의원') & (gender['제안정당'] == '')]
```

```python
# 국회의원일 때 
gender.loc[(gender['발의자_한글'] == '홍익표'), '제안정당'] = '더불어민주당'
gender.loc[(gender['발의자_한글'] == '황인자'), '제안정당'] = '새누리당'
gender.loc[(gender['발의자_한글'] == '전정희'), '제안정당'] = '국민의당'
gender.loc[(gender['발의자_한글'] == '원혜영'), '제안정당'] = '더불어민주당'
gender.loc[(gender['발의자_한글'] == '이찬열'), '제안정당'] = '더불어민주당'
gender.loc[(gender['발의자_한글'] == '김현미'), '제안정당'] = '더불어민주당'

# 정부, 위원장 발의일 때 
gender.loc[gender['제안자구분'] == '정부', '제안정당'] = '정부'
gender.loc[gender['제안자구분'] == '정부', '발의자'] = '정부'
gender.loc[gender['제안자구분'] == '위원장', '제안정당'] = '위원장'
gender.loc[gender['제안자구분'] == '위원장', '발의자'] = '위원장'
```

## 발의자 성별 구분

```python
def get_proposer_list(name_list):
    new_list = []
    for name in name_list: 
        clean_name = name.replace('(',' ').replace(')',' ').replace('/',' ').split(' ')
        new_list.append(clean_name[0].strip())
    return new_list 
```

```python
%time gender['제안자_한글'] = gender['제안자'].apply(get_proposer_list)
```

```python
women_list = ['박근혜','나경원','김을동','김희정','정미경','박인숙','권은희',
             '민병주','윤명희','강은희','신의진','이에리사','김현숙','이자스민',
             '류지영','민현주','손인춘','신경림','문정림','황인자','박윤옥',
             '장정은','정윤숙','이미경','추미애','박영선','유승희','김영주',
             '김상희','김현미','서영교','인재근','박혜자','이언주','유은혜',
             '전순옥','은수미','진선미','배재정','남인순','한정애','장하나',
             '김현','최민희','임수경','권은희','전정희','권은희','전정희',
             '심상정','김제남']

gender.loc[gender['발의자_한글'].isin(women_list), '성별'] = '여성'
gender.loc[(gender['제안자구분'] == '의원') & (~gender['발의자_한글'].isin(women_list)), '성별'] = '남성'
```

```python
gender['성별'].unique()
```

```python
def count_women(name_list, women_list): 
    count = 0 
    for name in name_list: 
        if name in women_list: 
            count += 1
    return count
```

```python
%time gender['여성의원 수'] = gender['제안자_한글'].apply(lambda x: count_women(x, women_list))
```
