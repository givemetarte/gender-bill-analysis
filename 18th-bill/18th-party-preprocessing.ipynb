```python
import pandas as pd 
import numpy as np 
from ast import literal_eval
```

## 발의자/제안정당

```python
%time gender['제안자'] = gender['제안자'].apply(literal_eval)
```

```python
def get_proposer(name_list): 
    if len(name_list) > 0: 
        proposer = name_list[0]
    else: 
        proposer = ''
    return proposer
```

```python
%time gender['발의자'] = gender['제안자'].apply(get_proposer)
```

## 발의자 성별 구분

```python
women_list = ['박근혜','김영선','전재희','나경원','진수희','전여옥','이혜훈',
             '박순자','박영아','정미경','강명순','배은희','이정선','김소남',
             '이은재','김금래','조윤선','손숙미','이애주','정옥임','김옥이',
             '이두아','최경희','이영애','이미경','추미애','조배숙','박영선',
             '신낙균','이성남','최영희','전혜숙','전현희','김상희','박선숙',
             '김유정','김진애','송영선','김을동','정영희','김혜성','김정',
             '이영애','박선영','곽정숙','이정희']

gender.loc[gender['발의자'].isin(women_list), '성별'] = '여성'
gender.loc[(gender['제안자구분'] == '의원') & (~gender['발의자'].isin(women_list)), '성별'] = '남성'
```

```python
gender['성별'].unique()
```

```python
def count_women(name_list, women_list): 
    count = 0 
    for name in name_list: 
        if name in women_list: 
            count += 1
    return count
```

```python
%time gender['여성의원 수'] = gender['제안자'].apply(lambda x: count_women(x, women_list))
```
