## 발의자 성별 구분

```python
women_list = ['이미경', '김희선', '한명숙', '조배숙','김선미', '장향숙', '김명자', '이경숙', '홍미영', '박영선', '김현미', '김영주', '강혜숙', '이은영', '윤원호', '유승희', '장복심', '서혜석', '신명',
             '박근혜','김영선','전재희','이혜훈','김희정','김애실','박찬숙','송영선','전여옥','이계경','나경원','김영숙','고경화','진수희','안명옥','박순자','문희', 
             '심상정','이영순','최순영','현애자','손봉숙','이승희','김송자']

gender.loc[gender['발의자'].isin(women_list), '성별'] = '여성'
gender.loc[(gender['제안자구분'] == '의원') & (~gender['발의자'].isin(women_list)), '성별'] = '남성'
```

```python
gender['성별'].unique()
```

```python
def count_women(name_list, women_list): 
    count = 0 
    for name in name_list: 
        if name in women_list: 
            count += 1
    return count
```

```python
%time gender['여성의원 수'] = gender['제안자'].apply(lambda x: count_women(x, women_list))
```
